```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10, 0.01)
print(x)
print(x.size)
```

```python
def mse_loss(ypred, y):
  return np.sum((ypred-y)**2)/y.size

def mae_loss(ypred, y):
  return np.sum(np.abs(ypred-y))/y.size

def huber_loss(ypred, y, delta=1):
  huber_mse = 0.5*np.sum((ypred-y)**2)/y.size
  huber_mae = delta*(np.sum(np.abs(ypred-y))-0.5*delta)
  return np.where(np.abs(ypred-y)<=delta, huber_mse, huber_mae)

x = np.arange(-5, 5, 0.01)

# MSE plot
y_square = np.square(x)

# MAE plot
y_abs = np.abs(x)

# Huber plot
delta=1.5
y_huber = np.where(np.abs(x)<delta, 0.5*np.square(x), delta*(np.abs(x)-0.5*delta))

plt.figure(figsize=(12, 6))
plt.plot(x, y_square, label="MSE")
plt.plot(x, y_abs, label="MAE")
plt.plot(x, y_huber, label="Huber")
plt.axis([-4, 4, 0, 20])
plt.grid(True)
plt.legend()
plt.show()
```
